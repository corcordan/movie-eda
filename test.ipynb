```python
import pandas as pd

def unwrap_quotes(title: str) -> str:
    return title[1:-1] if title.startswith('"') and title.endswith('"') else title

test = pd.Series([
    '"ABCDE"',
    'AB-"CDE"',
    '"The "Great" Escape"',
    '"Hello"',
    'Normal Title'
])

test = test.apply(unwrap_quotes)
print(test)
```
